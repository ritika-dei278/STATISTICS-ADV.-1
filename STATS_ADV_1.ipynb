```python
#2  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)>
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.arange(1, 7)  # Example: Rolling a fair die (values 1 to 6)
pmf = stats.randint.pmf(x, 1, 7)

plt.bar(x, pmf, color='blue', alpha=0.6)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution")
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

p = 0.5  # Probability of success
print("PDF for x=1:", bernoulli_pdf(p, 1))
print("PDF for x=0:", bernoulli_pdf(p, 0))
```

```python
#16 Create a confidence interval for a dataset using Python and interpret the result5
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean, std = np.mean(data), np.std(data, ddof=1)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std/np.sqrt(len(data)))

print("95% Confidence Interval:", ci)
```

```python
#17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean5
import numpy as np
import scipy.stats as stats

np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, SD=15, Sample Size=50

mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # Using sample standard deviation

confidence_level = 0.95  # 95% Confidence Interval
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)  # Z-score for 95% confidence

margin_of_error = z_critical * (std_dev / np.sqrt(len(data)))
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

```
